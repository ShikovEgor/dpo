# Базовые типы данных в Python 

## 1. Базовые типы: численный

```python
# int
num = 5
print(num)
print(type(num))
```

```python
# float
num = 13.4
print(num)

num = 1.5e2
print(num)
```

```python
# Python may convert types
num = int(num)
print(num, type(num))

num = float(num)
print(num, type(num))
```

## 2. Базовые типы: логический

```python
# bool
print(3 > 4)
print(3 <= 3)
print(6 >= 6)
print(6 < 5)
```

```python
x=False
y=True
print(x and y) # logical and
print(x or y) # logical or
print(not y) # logical not
```

```python
# example
x, y, z = True, False, True
result = x and y or z
print(result)
```

## 4*. Множества

```python
empty_set = set()
number_set = {1, 2, 3, 3, 3, 3, 4, 5}

print(number_set)
```

```python
odd_set, odd_set = set(), set()

odd_set = {1, 3, 5, 7}
even_set = {2, 4, 6, 8}

print(odd_set)
print(even_set)

union_set = odd_set | even_set
print(union_set)

intersection_set = odd_set & even_set
print(intersection_set)
```

## Описание:
#### Данные клиента:
1 - age: возраст<br>
2 - job: тип работы (админ., синий воротничок, предприниматель, горничная, менеджмент, пенсионер, частный предприниматель, сфера услуг, студент, техник, безработный, неизвестно)<br>
3 - marital: семейное положение (разведен, женат, не женат, неизвестно; Примечание: "разведен" означает разведен или овдовел)<br>
4 - education: образование(начальное, среднее, высшее и неизвестное)<br>
5 - default:  имеет кредит по умолчанию? (нет, да, неизвестно.)<br>
6 - housing: есть ли ипотека? (нет, да, неизвестно.)<br>
7 - loan: есть ли личный  кредит? (нет, да, неизвестно.)<br>
8 - balance: баланс.

#### Данные по последнему контакту в рамках кампании:
9 - contact: тип контакта ("сотовая связь", "телефон") <br>
10 - month: последний контактный месяц ('jan', 'feb', 'mar', ..., 'nov', 'dec')<br>
11 - day: последний контактный день недели ('mon','tue','wed','thu','fri')<br>
12 - длительность последнего контакта, в секундах <br>

#### Прочие:
13 - campaign: количество контактов, выполненных во время этой кампании и для данного клиента (включает последний контакт)<br>
14 - pdays: количество дней, прошедших после последнего обращения к клиенту из предыдущей кампании (999 означает, что клиент ранее не обращался)<br>
15 - previous: количество контактов, выполненных до данной кампании для данного клиента <br>
16 - poutcome: результат предыдущей маркетинговой кампании ("неудача", "несуществующий", "успех"))<br>

#### Целевая переменная (желаемая цель):
17 - клиент оформил срочный депозит? ("да", "нет")

## 3. Кластеризация 
<bar>

```python
# Performing one-hot encoding for all the categorical columns
df_dig = pd.get_dummies(df.drop(columns = ['deposit']))
df_dig.head()
```

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=5, random_state=0).fit(df_dig)
labels = kmeans.labels_
```

```python
from sklearn.manifold import TSNE
tsne = TSNE(n_components = 2, init = 'pca',)
data_2d_tsne = tsne.fit_transform(df_dig)
```

```python
plt.figure(figsize = (10, 6))
plt.scatter(data_2d_tsne[:, 0], data_2d_tsne[:, 1], c = labels)
```
